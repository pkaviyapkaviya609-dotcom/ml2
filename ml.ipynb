```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Redefine x and y to ensure they are up-to-date and correctly preprocessed
x = df.iloc[:,:-1]
x = pd.get_dummies(x, drop_first=True) # Apply one-hot encoding to categorical features
y = df.iloc[:,-1]

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42) # Added random_state for reproducibility
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

clf = RandomForestClassifier(random_state=42)
clf.fit(xtrain, ytrain)

ypred = clf.predict(xtest)

accuracy = accuracy_score(ytest, ypred)
print(accuracy)
```

```python
import pickle
with open('model.pkl', 'wb') as file:
    pickle.dump(clf, file)
```
